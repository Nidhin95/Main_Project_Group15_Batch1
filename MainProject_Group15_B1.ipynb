# Exploratory Data Analysis

```python
# Distribution graphs (histogram/bar graph) of column data (Univariate)
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique()
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 50]] # For displaying purposes, pick columns that have between 1 and 50 unique values
    nRow, nCol = df.shape
    columnNames = list(df)
    nGraphRow = (nCol + nGraphPerRow - 1) / nGraphPerRow
    plt.figure(num = None, figsize = (2 * nGraphPerRow, 8 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{columnNames[i]} ')
    plt.tight_layout(pad = 0.8, w_pad = 5.0, h_pad = .25)
    plt.show()
    
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
plotPerColumnDistribution(d, 20, 4)
```

# State

```python
d.groupby('State')['Customer Lifetime Value'].mean()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
d.State.value_counts().plot.pie(shadow=True,autopct='%1.1f%%',ax=ax[0],colors =['teal','lightseagreen','mediumaquamarine','palegreen','cyan'])

sns.barplot(x='State',y="Customer Lifetime Value",palette="Paired", edgecolor=".6",data=d,ax=ax[1])
plt.title('States vs Customer Lifetime Value ')
plt.show()
```

**1. 34.5% of customers are from California.**

**2. Customers from Washington are comparitively less in number.**

**3. Average customer lifetime value is low for the customers from Arizona.**

**4. Customers from all other states have almost equal average customer lifetime value.**

# Response

```python
d.groupby('Response')['Customer Lifetime Value'].mean()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
axi=sns.countplot(x='Response',palette="viridis",data=d,ax=ax[0])
plt.title('Response')
for p in axi.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/float(len(d)))
    x = (p.get_x() + p.get_width())/1.23
    y = p.get_height()
    axi.annotate(percentage, (x, y),ha='center')

   

sns.barplot(x='Response',y="Customer Lifetime Value",palette="viridis", edgecolor=".6",data=d,ax=ax[1])
plt.title('Response vs Customer Lifetime Value')
```

**1. 85.7% of customers's don't like to renew their insurance .**

**2. Average customer lifetime value is slightly higher for customers who don't like to renew their insurance. So it is important to retain those customers .**
